```python
import numpy as np
from sklearn.cluster import KMeans
```

Step 1: Load data from places.txt
Each line contains a longitude-latitude pair in the format "-112.1,33.5"

```python
with open("places.txt", "r") as file:
    locations = [list(map(float, line.strip().split(","))) for line in file]
print(locations)
```

```python
locations = np.array(locations)
```

Step 3:Apply K-Means clustering with k=3


```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(locations)
```

Step 4:Write output to clusters.txt

```python
with open("clusters.txt", "w") as file:
    for idx, label in enumerate(labels):
        file.write(f"{idx} {label}\n")
```

Step 5:Print completion message

```python
print("Clustering complete. Output saved in clusters.txt")
```

Step 7: Print the output file

```python
with open("clusters.txt", "r") as file:
    for line in file:
        print(line.strip()) 
```
